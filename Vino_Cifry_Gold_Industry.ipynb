### Проверка эффективности обогащения

Нужно проверить, верно ли подсчитана эффективность обогащения (столбец rougher.output.recovery). Для этого у меня есть формула рассчёта эффективности обогащения. Для рассчёта буду использовать полный датасет.

```python
#доля золота в концентрате после флотации
c = data_raw_train['rougher.output.concentrate_au']
#доля золота в сырье до флотации
f = data_raw_train['rougher.input.feed_au']
#доля золота в хвостах после флотации
t = data_raw_train['rougher.output.tail_au']

calculated_recovery = (c * (f - t)) / (f * (c - t)) * 100
calculated_recovery.head(3)
```

Я рассчитала по формуле рассчётную эффективность, теперь сравню её с фактической используя метрику MAE.

```python
print('MAE фактической эффективности флотации и рассчётной равно:',
      mean_absolute_error(data_raw_train['rougher.output.recovery'], calculated_recovery))
```

MAE получилось равным 9,8 что для рассчёта эффективности, где значения могут быть до 100 (процентов), приемлемо.

### Функции рассчёта качества

Напишу функции, одна из которых будет считать sMAPE, а другая суммировать их в итоговое sMAPE.

```python
# функция sMAPE
def calculate_smape(target, pred):
    return (1 / len(target)) * np.sum(np.abs(target - pred) / (np.abs(target + pred) / 2) * 100)
```

```python
# функция для итогового sMAPE
def total_smape(smape_r, smape_f):
    return (0.25 * smape_r) + (0.75 * smape_f)
```

```python
#создание scoring для cross_val_score
scoring_smape = make_scorer(calculate_smape, greater_is_better=False)
```

Функция total_smape значения sMAPE целевого признака 'rougher.output.recovery' и 'final.output.recovery'. Возвращает сумму их произведений на 25 и 75 процентов соответственно.

### Регрессионное решающее дерево

Обучаю модель решающее дерево. Оценю её при помощи метода кросс-валидации, сперва на одном целевом признаке, затем на другом. Получу средние оценки и суммирую их.

```python
model_tree = None
best_total_score = 1000
best_depth_tree = 0
for depth in range(1, 10):
    model = DecisionTreeRegressor(random_state=12345, max_depth=depth)
    score_rougher = abs(cross_val_score(model, features_train, target_rougher_train, scoring=scoring_smape).mean())
    score_final = abs(cross_val_score(model, features_train, target_final_train, scoring=scoring_smape).mean())
    score = total_smape(score_rougher, score_final)
    if score < best_total_score:
        model_tree = model
        best_total_score = score
        best_depth_tree = depth

print("Лучшее общее sMAPE кросс-валидации модели решающего дерева:", best_total_score, "Глубина дерева:", best_depth_tree)
```

Лучшее дерево с глубиной 2 дало среднюю оценку sMAPE 11,56.

### Регрессионный случайный лес
